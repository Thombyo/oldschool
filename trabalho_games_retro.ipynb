```python
login_stack = []

def login(user):
    login_stack.append(user)
    print(f"{user} fez login.")

def logout():
    if login_stack:
        user = login_stack.pop()
        print(f"{user} fez logout.")
    else:
        print("Nenhum usuário logado.")

while True:
    acao = input("Digite 'login' para login, 'logout' para logout, ou 'sair' para sair: ").lower()
    if acao == 'login':
        nome_usuario = input("Digite o nome do usuário para login: ")
        login(nome_usuario)
        break
    elif acao == 'logout':
        logout()
    elif acao == 'sair':
        print("Saindo do login.")
        exit()
    else:
        print("Comando inválido. Por favor, tente novamente.")

import numpy as np
produtos_array = np.array([
    "punch out", "the legend of zelda", "pokemon red", "pokemon blue",
    "pokemon yellow", "pokemon crystal", "pokemon gold", "pokémon silver",
    "pokemon green", "super mario bros", "doom", "street fighter", "duck hunt",
    "excitebike", "castlevania", "pac-man", "space invaders", "galaga",
    "donkey kong", "detetive pikachu"
])

def listar_produtos_array(array_produtos):
    print("Produtos disponíveis:")
    for i, produto in enumerate(array_produtos):
        print(f"{i + 1}. {produto}")

listar_produtos_array(produtos_array)

carrinho = []
while True:
    acao = input("Digite '1' para adicionar e '2' para sair: ")
    if acao == '1':
        adicionar = input("Adicione um jogo ao carrinho: ").lower()
        if adicionar in produtos_array:
            carrinho.append(adicionar)
            print(f"{adicionar} adicionado ao carrinho.")
        else:
            print("Produto não encontrado.")
    elif acao == '2':
        print("Encerrado")
        print("Carrinho:", carrinho)
        break
    else:
        print("Comando inválido.")

metodos_pagamento = [
    ["Cartão de crédito", "Boleto", "Pix"],
    [5.0, 2.0, 0.0]
]

def mostrar_metodos_pagamento():
    print("Métodos de Pagamento:")
    for metodo, taxa in zip(metodos_pagamento[0], metodos_pagamento[1]):
        print(f"{metodo} - Taxa: {taxa}%")

mostrar_metodos_pagamento()

while True:
    metodo_pagamento = input("Digite o método desejado: ")
    if metodo_pagamento in metodos_pagamento[0]:
        print("O método escolhido foi:", metodo_pagamento)
        break
    else:
        print("Método inválido. Tente novamente.")

class Fila:
    def __init__(self):
        self.fila = []

    def adicionar(self, nome_usuario, carrinho, metodo_pagamento):
        self.fila.append({
            "nome": nome_usuario,
            "carrinho": carrinho,
            "metodo": metodo_pagamento
        })
        print("Usuário adicionado à fila.")

    def login(self):
        if self.fila:
            usuario = self.fila.pop(0)
            print(f"Logando usuário: {usuario['nome']}")
            return usuario
        else:
            print("Fila vazia.")
            return None

    def mostrar(self):
        if self.fila:
            print("Usuários na fila:")
            for u in self.fila:
                print(f"Usuário: {u['nome']}, Pagamento: {u['metodo']}, Carrinho: {u['carrinho']}")
        else:
            print("Fila vazia.")

fila = Fila()
fila.adicionar(nome_usuario, carrinho, metodo_pagamento)

fila.mostrar()
]
usuario_logado = fila.login()
print("Usuário logado:", usuario_logado)
```
